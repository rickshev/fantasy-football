```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = pd.read_csv("Data/fantasy.csv")
qb = pd.read_csv("Data/qb.csv")
rb = pd.read_csv("Data/rb.csv")
wr = pd.read_csv("Data/wr.csv")
te = pd.read_csv("Data/te.csv")
qb.head()
```

```python
grouped_qb = qb.groupby("Value")
grouped_qb.count()
```

```python
X = qb.drop(["Player", "Tm", "FantPos", "PosRank", "OvRank", "Season", "Value"], axis=1)
y = qb["Value"]
prediction_labels = qb["Value"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
import tensorflow
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=1, stratify=y)
```

```python
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
# Step 2: Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
# Creating Deep Learning Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Create model and add layers
model = Sequential()
model.add(Dense(units=100, activation='relu', input_dim=28))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=5, activation='softmax'))
```

```python
# Compile and fit the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=60,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```
